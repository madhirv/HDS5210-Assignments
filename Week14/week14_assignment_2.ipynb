### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
"""
Data Analysis Script
This script uploads a CSV file, calculates the number of unique values in each column,
identifies the column with the most unique values, and determines whether it's a unique
identifier. If it is, it also identifies the column with the second-highest number of
unique values.
"""

# Step 1: Importing necessary libraries
import pandas as pd
from google.colab import files

# Step 2: Uploading the CSV file
print("Upload your file...")
uploaded = files.upload()  # Opens file upload dialog in Colab

# Step 3: Loading the CSV file into a DataFrame
"""
Loads the uploaded CSV file into a Pandas DataFrame for analysis.The file name is dynamically extracted from the uploaded files.
"""
file_name = list(uploaded.keys())[0]
data = pd.read_csv(file_name)

# Step 4: Calculating the number of unique values in each column
"""
Calculates the number of unique values for each column in the dataset and displays the results.

"""
print("\nNumber of unique values in each column:")
unique_counts = data.nunique()
print(unique_counts)

# Step 5: Identifying the column with unique values
"""
Identifying the column with the highest number of unique values and displays the column name and the corresponding count of unique values.
"""
most_unique_column = unique_counts.idxmax()
most_unique_count = unique_counts.max()
print(f"\nColumn with the most unique values: '{most_unique_column}' ({most_unique_count} unique values)")

# Step 6: If the column is a unique identifier, finding the second-highest
"""
Checking if the column with the most unique values is a unique identifier  If so, it identifies the column with the second-most
unique values.
"""
if most_unique_count == len(data):
    print(f"'{most_unique_column}' appears to be a unique identifier.")
    # Finding the second-most unique column
    second_most_unique_column = unique_counts.drop(most_unique_column).idxmax()
    second_most_unique_count = unique_counts.drop(most_unique_column).max()
    print(f"Column with the second most unique values: '{second_most_unique_column}' ({second_most_unique_count} unique values)")

```

### 3. Data Redundancy

Use Python to identify any columns in your data where the value for every row is the same.  For example, if you had a gender column and all the rows said 'M', then you would want to write the code that helped you determine that.

```python
# Importing the necessary libraries
import pandas as pd
from google.colab import files

def check_redundant_columns():
    """
    This function allows the user to upload a CSV file,
    checks for any redundant columns in the dataset,
    and displays the results.

    Steps done here are-
    1. Upload a CSV file using Google Colab's file upload dialog.
    2. Load the CSV file into a Pandas DataFrame.
    3. Iterate through each column in the dataset to check for redundancy.
    4. Display the redundant columns or indicate if no redundant columns exist.
    """
    # Step 1: Uploading your dataset
    print("Upload your CSV file...")
    uploaded_csv = files.upload()

    # Step 2: Loading the uploaded file into a Pandas DataFrame
    csv_file_name = list(uploaded_csv.keys())[0]
    csv_data = pd.read_csv(csv_file_name)

    # Step 3: Checking for redundant columns and a column is redundant if all its values are the same.
    redundant_columns = []
    for column in csv_data.columns:
        if csv_data[column].nunique() == 1:
            redundant_columns.append(column)

    # Step 4: Displaying results
    if redundant_columns:
        print("Redundant Columns (all rows have the same value):")
        print(redundant_columns)
    else:
        print("No redundant columns found in the dataset.")

# Calling the function
check_redundant_columns()

```

```python
"""The result "No redundant columns found in the dataset" indicates that all columns in the uploaded dataset contain more than one unique value, meaning there are no columns where every row has the exact same value. This suggests that the data in each column is diverse and likely useful for analysis, as every column contributes unique information to the dataset. The code accomplishes this by iterating through each column, checking the number of unique values using the `.nunique()` method, and determining if any column contains only one unique value. Since no such columns were identified, the dataset is well-structured and free from redundancy."""
```

```python
"""
This script generates histograms for the 'Anxiety disorders (%)' and
'Alcohol use disorders (%)' columns from a CSV dataset.
It uses Matplotlib for visualization and Pandas for data manipulation.
"""

# Step 1: Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Step 2: Uploading the CSV file
print("Upload your CSV file...")
uploaded = files.upload()

# Step 3: Loading the CSV file into a DataFrame
file_name = list(uploaded.keys())[0]
data = pd.read_csv(file_name)

# Step 4: Ensuring numeric conversion for necessary columns and converts the 'Year' column to numeric format and drops rows with invalid or missing years
data['Year'] = pd.to_numeric(data['Year'], errors='coerce')
data = data.dropna(subset=['Year'])

# Step 5: Creating histograms for the selected numeric columns
plt.figure(figsize=(12, 6))

# Histogram for Anxiety disorders (%)
plt.subplot(1, 2, 1)
plt.hist(data['Anxiety disorders (%)'].dropna(), bins=30, edgecolor='black', color='blue')
plt.title('Histogram of Anxiety Disorders (%)')
plt.xlabel('Anxiety Disorders (%)')
plt.ylabel('Frequency')

# Histogram for Alcohol use disorders (%)
plt.subplot(1, 2, 2)
plt.hist(data['Alcohol use disorders (%)'].dropna(), bins=30, edgecolor='black', color='orange')
plt.title('Histogram of Alcohol Use Disorders (%)')
plt.xlabel('Alcohol Use Disorders (%)')
plt.ylabel('Frequency')

# displays the plots
plt.tight_layout()
plt.show()


```
