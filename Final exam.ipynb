```python
from google.colab import files
import pandas as pd

# Step 1: Uploading the datasets using Google Colab's file upload utility
"""
Uploading files interactively in Google Colab.
"""
uploaded = files.upload()

# Step 2: Defining a function to load and explore datasets
"""
Function to load and display basic information about a dataset.

Workflow:
    - Loads the dataset into a Pandas DataFrame based on the specified file type.
    - Displays the first few rows of the dataset (`head()`).
    - Prints column names to identify dataset structure.
    - Shows dataset summary information (`info()`).

"""
def process_dataset(file_name, file_type='csv', encoding=None):
    print(f"\n--- Processing {file_name} ---")
    try:
        # Loading the dataset based on file type
        if file_type == 'csv':
            # Loading CSV file
            data = pd.read_csv(file_name, encoding=encoding)
        elif file_type == 'json':
            # Loading JSON file
            data = pd.read_json(file_name, orient='records', lines=True)
        elif file_type == 'txt':
            # Loading TXT file with tab-separated values
            data = pd.read_csv(file_name, sep='\t', encoding=encoding)
        else:
            # Handling the unsupported file types
            print(f"Unsupported file type for {file_name}")
            return

        # Displaying success message and dataset structure
        print(f"Successfully loaded {file_name}")
        print(f"Columns in {file_name}:\n", data.columns)
        print(data.head())  # Display first few rows for a quick preview
        print(data.info())  # Show detailed dataset information
        return data
    except Exception as e:

        print(f"Error processing {file_name}: {e}")

# Step 3: Processing each dataset individually and store in a dictionary
"""
Dictionary holds all processed datasets with their file names as keys.
This ensures datasets are accessible for further analysis.
"""
datasets = {}

# Processing each dataset with specific details for file type and encoding
"""
The datasets represent various environmental, agricultural, and disaster-related data.
Each dataset is loaded using the `process_dataset` function.
"""
datasets['Environment_Temperature_change'] = process_dataset(
    "Environment_Temperature_change_E_All_Data_NOFLAG.csv", file_type='csv', encoding='ISO-8859-1'
)  # Temperature data with special encoding

datasets['FAOSTAT_2022'] = process_dataset(
    "FAOSTAT_data_1-10-2022.csv", file_type='csv'
)  # Agricultural dataset from 2022

datasets['FAOSTAT_2020'] = process_dataset(
    "FAOSTAT_data_11-24-2020.csv", file_type='csv'
)  # Agricultural dataset from 2020

datasets['FAOSTAT_2024'] = process_dataset(
    "FAOSTAT_data_en_11-1-2024.csv", file_type='csv'
)  # Agricultural dataset from 2024

datasets['Greenhouse_Emissions'] = process_dataset(
    "Green_house_emissions.json", file_type='json'
)  # Greenhouse gas emissions in JSON format

datasets['US_Disaster_Declarations'] = process_dataset(
    "us_disaster_declarations.txt", file_type='txt'
)  # Disaster declarations in TXT format with tab-separated values

datasets['US_Disasters_M5'] = process_dataset(
    "us_disasters_m5.txt", file_type='txt'
)  # Additional disaster data in TXT format
```

Data Merging

Data merging is a critical step in this project as it brings together multiple datasets to create a comprehensive and unified view of global environmental and disaster trends. For this project, we utilized seven datasets, each containing unique information relevant to the analysis. By merging these datasets, we can analyze trends across multiple dimensions, such as regions, years, and domains.

Datasets Used for Merging:
1. Greenhouse Gas Emissions Dataset: Contains data on CO2 emissions and other greenhouse gases by region and year.
2. FAOSTAT Data (three separate files): Provides information on temperature changes and related climate data for different countries and regions over time.
3. US Disaster Declarations Dataset: Includes information on disaster events in the United States, categorized by region, type of disaster, and declaration date.
4. US Disasters M5 Dataset: Another dataset detailing disaster metrics, providing additional regional and programmatic insights.
5. Environmental Temperature Change Dataset: Contains temperature change data for regions over several decades.


Merging Process:

The merging process involved standardizing column names across datasets (e.g., ensuring columns like region and year are consistent).

Specific columns such as region and year were used as keys to perform inner joins, ensuring only relevant data with common values across datasets were retained.

For datasets without explicit year or region columns, derived attributes (e.g., extracting year from the declaration_date) were created to enable seamless merging.

After merging, the resulting dataset contains attributes like greenhouse gas emissions, temperature changes, and disaster declarations, aligned by region and year.

This merged dataset forms the foundation for further analysis, allowing us to aggregate, pivot, and visualize data for meaningful insights into environmental trends and disaster impacts globally.

```python
from google.colab import files
import pandas as pd

"""
This script facilitates the merging of multiple datasets for analysis.
It includes functionality for uploading datasets, loading them into pandas DataFrames with encoding fixes,
standardizing columns for merging, and finally saving the merged dataset for further analysis.
"""

# Step 1: Uploading Files
print("Please upload your datasets:")
uploaded_files = files.upload()

# Step 2: Loading Files into DataFrames with Encoding Fix
"""
Load the uploaded files into pandas DataFrames, accounting for different file formats
such as CSV, TXT, and JSON. Also handles encoding issues specific to certain datasets.
"""
dataframes = {}

for file_name in uploaded_files.keys():
    try:
        if file_name.endswith('.csv'):
            # Fix encoding for specific files
            if 'Environment_Temperature_change_E_All_Data_NOFLAG' in file_name:
                df = pd.read_csv(file_name, encoding='latin1')
            else:
                df = pd.read_csv(file_name)
            print(f"'{file_name}' loaded successfully with shape {df.shape}")

        elif file_name.endswith('.txt'):
            # Loading TXT file (assume tab-delimited)
            df = pd.read_csv(file_name, delimiter='\t')
            print(f"'{file_name}' loaded successfully with shape {df.shape}")

        elif file_name.endswith('.json'):
            # Loading JSON file
            df = pd.read_json(file_name, lines=True)
            print(f"'{file_name}' loaded successfully with shape {df.shape}")

        else:
            # Notify if unsupported file format is encountered
            print(f"Unsupported file format: {file_name}")
            continue

        # Storing the DataFrame in the dictionary with the file name as the key
        dataframes[file_name] = df

    except Exception as e:

        print(f"Error loading file {file_name}: {e}")

# Step 3: Columns for Merging
"""
Standardize column names and prepare datasets for merging. This ensures consistency
in column names like 'region' or 'state' for effective merging across datasets.
"""
# Getting dataframes for specific datasets
df_emissions = dataframes.get('Green_house_emissions (5).json')
df_disasters = dataframes.get('us_disaster_declarations (3).txt')

# Standardize columns for merging
if df_emissions is not None and df_disasters is not None:
    # Ensure consistency in 'Region' and 'state' columns by stripping whitespace and converting to uppercase
    df_emissions['Region'] = df_emissions['Region'].str.strip().str.upper()
    df_disasters['state'] = df_disasters['state'].str.strip().str.upper()

    # Merging datasets on 'Region' and 'state'
    merged_data = pd.merge(df_emissions, df_disasters, left_on='Region', right_on='state', how='inner')
    print(f"Merged dataset shape: {merged_data.shape}")

    # Saving the merged dataset
    merged_data.to_csv('merged_emissions_disasters.csv', index=False)
    print("Merged dataset saved as 'merged_emissions_disasters.csv'.")
else:
    # Notify if one or both datasets are missing for merging
    print("One or both datasets required for merging are missing.")

# Step 4: Inspect DataFrames
"""
Iterate through all uploaded DataFrames to display their structure and the first few rows.
This helps in verifying the integrity and content of the datasets.
"""
for file_name, df in dataframes.items():
    print(f"\nSummary for '{file_name}':")
    print(df.info())  # Display the structure of the dataset
    print(df.head())  # Display the first few rows of the dataset
```

```python
"""
Data Aggregation Step for Climate and Disaster Analysis Project

This step involves calculating aggregated metrics across multiple datasets
to extract meaningful insights and prepare the data for visualization and reporting.

Steps:
1. Aggregate data based on `region` and `year` columns.
2. Compute summary statistics (e.g., mean, sum, count).
3. Add derived columns (e.g., value ratios or normalized metrics).
4. Save aggregated results for further analysis.
"""

# Step 1: Importing required libraries
import pandas as pd

# Step 2: Loading final dataset
print("\n--- Loading Final Dataset ---")
try:
    final_data = pd.read_csv('final_dataset.csv')
    print("Final dataset loaded successfully.")
    print(f"Final dataset shape: {final_data.shape}")
    print(final_data.head())
except FileNotFoundError:
    print("Error: 'final_dataset.csv' not found. Ensure the merging step is completed.")
    raise

# Step 3: Performing aggregation on the dataset
print("\n--- Aggregating Data ---")
try:
    # Verifying columns in final_data
    print("Columns in final_data before aggregation:", final_data.columns)

    # Renaming 'Area' to 'region' for consistency
    if 'Area' in final_data.columns:
        final_data.rename(columns={'Area': 'region'}, inplace=True)
        print("'Area' column renamed to 'region'.")

    # Aggregating data by 'region' and 'year' with multiple metrics
    aggregated_data = final_data.groupby(['region', 'Domain']).agg(
        {
            'Value': ['sum', 'mean', 'count'],
            'Y1961': 'mean',
        }
    )

    # Saving the aggregated data
    aggregated_data.to_csv('aggregated_dataset.csv')
    print("Aggregated dataset saved as 'aggregated_dataset.csv'.")

    # Displaying a sample of the aggregated data
    print("\nAggregated Data Preview:")
    print(aggregated_data.head())

except KeyError as e:
    print(f"Error in aggregation: Missing column {e}")
except Exception as e:
    print(f"Unexpected error during aggregation: {e}")
```
